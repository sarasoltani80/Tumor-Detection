```python
import cv2
import os

image_dir = "/content/drive/My Drive/resized3"
output_dir = "/content/drive/My Drive/blurred_images3/"

os.makedirs(output_dir, exist_ok=True)

# Loop through each image in the directory
for image_name in os.listdir(image_dir):
    img_path = os.path.join(image_dir, image_name)

    img = cv2.imread(img_path)

    # Check if the image was loaded successfully
    if img is None:
        print(f"Failed to load image: {img_path}")
        continue

    # Apply Gaussian blur
    blurred_img = cv2.GaussianBlur(img, (5, 5), 0)

    cv2.imwrite(os.path.join(output_dir, image_name), blurred_img)

    print(f"Blurred and saved: {image_name}")
```

```python
import cv2
import os
import numpy as np


image_dir = "/content/drive/My Drive/blurred_images3/"
output_dir = "/content/drive/My Drive/sharpened_images3/"


os.makedirs(output_dir, exist_ok=True)

# Define a sharpening kernel for the high-pass filter
high_pass_kernel = np.array([[-1, -1, -1],
                             [-1,  9, -1],
                             [-1, -1, -1]])

# Loop through each image in the directory
for image_name in os.listdir(image_dir):
    img_path = os.path.join(image_dir, image_name)

    img = cv2.imread(img_path)

    if img is None:
        print(f"Failed to load image: {img_path}")
        continue

    # Apply high-pass filter using convolution with the kernel
    sharpened_img = cv2.filter2D(img, -1, high_pass_kernel)

    cv2.imwrite(os.path.join(output_dir, image_name), sharpened_img)

    print(f"Sharpened and saved: {image_name}")
```

```python
import cv2
import numpy as np
import os

image_dir = "/content/drive/My Drive/sharpened_images3/"
output_dir_opening = "/content/drive/My Drive/dilated3/"


os.makedirs(output_dir_opening, exist_ok=True)

# Kernel for erosion and dilation
kernel = np.ones((5, 5), np.uint8)

# Loop through each image in the directory
for image_name in os.listdir(image_dir):
    img_path = os.path.join(image_dir, image_name)

    img = cv2.imread(img_path, 0)


    if img is None:
        print(f"Failed to load image: {img_path}")
        continue

    # Apply erosion first
    eroded_img = cv2.erode(img, kernel, iterations=1)

    # Apply dilation after erosion
    opened_img = cv2.dilate(eroded_img, kernel, iterations=1)

    cv2.imwrite(os.path.join(output_dir_opening, image_name), opened_img)

    print(f"Opening (erosion followed by dilation) applied and saved for: {image_name}")
```
